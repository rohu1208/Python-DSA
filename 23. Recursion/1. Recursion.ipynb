```python
def n_to_1(n):
    if n < 1:
        return # Parameterized Recursion
    print(n)
    n_to_1(n-1)
    
n_to_1(5)
```

```python
# Functional Recursion -- return something
# 1. Make a flow 
# 2. BAse Condition
# 3. Recursive Call
def sum_n(num):
    # 1. Make a flow
    if num == 0:
        return 0 # 2. Base Condition
    else:
        return num + sum_n(num - 1)
sum_n(4)
```

```python
# BACKTRAKING

def funcc(i,n):
    if i<1:
        return 
    funcc(i-1,n)
    print(i)
funcc(4,4
      )
```
